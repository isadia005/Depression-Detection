```python
from nltk.corpus import stopwords
from textblob import TextBlob
from textblob import Word

columns = ['tweet_text', 'retweet_text', 'like_text'] #all column which need pre-processing will be here in this list
```

```python
#Convert to lowercase
#df[['tweet_text', 'retweet_text', 'like_text']] = df[['tweet_text', 'retweet_text', 'like_text']].apply(lambda x: x.str.lower())
for col in columns:
    df[col] = df[col].apply(lambda x: " ".join(x.lower() for x in x.split()))

df.head()
```

```python
# Removing numerical values
#df[['tweet_text', 'retweet_text', 'like_text']] = df[['tweet_text', 'retweet_text', 'like_text']].replace('\d+', '', regex=True)
for col in columns:
    df[col] = df[col].replace('\d+', '', regex=True)

df.head()
```

```python
# Removing punctations
#df[['tweet_text', 'retweet_text', 'like_text']] = df[['tweet_text', 'retweet_text', 'like_text']].replace("[^\w\s]", " ", regex=True)

for col in columns:
    df[col] = df[col].str.replace("[^\w\s]"," ", regex=True)
    df[col] = df[col].str.replace(r"(\x23.* )+", " ", regex=True)

df.head()
```

```python
# Removing remaining punctations

#df[['tweet_text', 'retweet_text', 'like_text']] = df[['tweet_text', 'retweet_text', 'like_text']].replace('_, __, :, "', " " , regex=True)

for col in columns:
    df[col] = df[col].str.replace('_, __, :, "', " " , regex=True)

df.head()
```

```python
# Removing double space

#df[['tweet_text', 'retweet_text', 'like_text']] = df[['tweet_text', 'retweet_text', 'like_text']].replace("\s+", " ", regex=True)

for col in columns:
    df[col] = df[col].str.replace("\s+", " ", regex=True)

df.head()
```

```python
# Removing links

#df[['tweet_text', 'retweet_text', 'like_text']] = df[['tweet_text', 'retweet_text', 'like_text']].replace("http\S+", " ", regex=True)


for col in columns:
    df[col] = df[col].str.replace('http\S+', ' ', regex=True)

df.head()
```

# K-means Clustering

```python
#K-means implementation
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Create bag of words matrix
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(negative_df['Negative Tweet'].values)
```

```python
# Dimensionality reduction with PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X.toarray())
```

```python
# Perform k-means clustering
kmeans = KMeans(n_clusters=3, max_iter=100, random_state=42, verbose=True)
kmeans.fit(X_pca)
```

```python
# Check the convergence of the centroids
print(kmeans.cluster_centers_)
```

```python
# Assign cluster labels to original data
negative_df['cluster_label'] = kmeans.labels_
```

```python
# Get indices of tweets in each cluster
cluster0_indices = negative_df[(negative_df['cluster_label'] == 0) & (negative_df['Depression Level'] == 'Highly Depressed')].index.tolist()
cluster1_indices = negative_df[(negative_df['cluster_label'] == 1) & (negative_df['Depression Level'] == 'Slightly Depressed')].index.tolist()
cluster2_indices = negative_df[(negative_df['cluster_label'] == 2) & (negative_df['Depression Level'] == 'Lightly Depressed')].index.tolist()
```

```python
# Display 10 rows of each cluster with index numbers
print('Cluster 0 (Highly Depressed):\n', negative_df.loc[cluster0_indices][:10])
print('Cluster 1 (Slightly Depressed):\n', negative_df.loc[cluster1_indices][:10])
print('Cluster 2 (Lightly Depressed):\n', negative_df.loc[cluster2_indices][:10])
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob
from textblob import Word

nltk.download("stopwords")
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("omw-1.4")

def preprocess_text(tweets_df, column_name):
    """
    Preprocesses text data in a given column of a pandas DataFrame.
    """
    # Convert to lowercase
    tweets_df[column_name] =tweets_df[column_name].apply(lambda x: " ".join(x.lower() for x in x.split()))
    # Removing numerical values
    tweets_df[column_name] =tweets_df[column_name].str.replace("\d","")
    # Removing punctations
    tweets_df[column_name] =tweets_df[column_name].str.replace("[^\w\s]","")
    tweets_df[column_name] =tweets_df[column_name].str.replace(r"(\x23.* )+", "")
    tweets_df[column_name] =tweets_df[column_name].str.replace('_', '')
    tweets_df[column_name] =tweets_df[column_name].str.replace('__', '')
    # Removing double space
    tweets_df[column_name] =tweets_df[column_name].str.replace("\s+", " ")
    # Removing user
    tweets_df[column_name] =tweets_df[column_name].str.replace('(@[A-Za-z]+[A-Za-z0-9-_]+)', '') #remove twitted at
    # Removing links
    tweets_df[column_name] =tweets_df[column_name].str.replace('http\S+', '')
    # Removing small words which are less than given condition
    tweets_df[column_name] =tweets_df[column_name].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
    # STOPWORDS
    sw = stopwords.words("english")
    tweets_df[column_name] =tweets_df[column_name].apply(lambda x: " ".join(x for x in x.split() if x not in sw))
    # Lemmatization (forgot converted into forget)
    tweets_df[column_name] =tweets_df[column_name].apply(lambda x: " ".join([Word(x).lemmatize()]))
    tweets_df[column_name+"_tokens"] = tweets_df[column_name].apply(lambda x: TextBlob(x).words)
    # Frequency Analysis
    tweets_df[column_name+"_frequency"] = tweets_df[column_name].apply(lambda x: len(str(x).split(" ")))
    
    return tweets_df
```
